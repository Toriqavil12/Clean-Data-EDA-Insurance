## CEK OUTLIER

```python
numeric_cols = [
    'age',
    'bmi',
    'children',
    'charges'
]

def detect_outliers_iqr(df_clean, columns):
    outlier_summary = {}

    for col in columns:
        q1 = df_clean[col].quantile(0.25)
        q3 = df_clean[col].quantile(0.75)
        iqr = q3 - q1

        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr

        outliers = df_clean[(df_clean[col] < lower_bound) | (df_clean[col] > upper_bound)]

        outlier_summary[col] = {
            'lower_bound': lower_bound,
            'upper_bound': upper_bound,
            'outlier_count': outliers.shape[0]
        }

    return pd.DataFrame(outlier_summary).T
```

```python
outlier_summary = detect_outliers_iqr(df_clean, numeric_cols)
outlier_summary
```

```python
Q1 = df['bmi'].quantile(0.25)
Q3 = df['bmi'].quantile(0.75)
IQR = Q3 - Q1

lower_bmi = Q1 - 1.5 * IQR
upper_bmi = Q3 + 1.5 * IQR

bmi_outliers = df[
    (df['bmi'] < lower_bmi) | (df['bmi'] > upper_bmi)
]

bmi_outliers
```

```python
bmi_outliers_display = bmi_outliers.reset_index(drop=True)
bmi_outliers_display.index = bmi_outliers_display.index + 1

bmi_outliers_display
```

```python
Q1 = df['charges'].quantile(0.25)
Q3 = df['charges'].quantile(0.75)
IQR = Q3 - Q1

lower_chg = Q1 - 1.5 * IQR
upper_chg = Q3 + 1.5 * IQR

charges_outliers = df[
    (df['charges'] < lower_chg) | (df['charges'] > upper_chg)
]

charges_outliers
```

```python
charges_outliers_display = charges_outliers.reset_index(drop=True)
charges_outliers_display.index = charges_outliers_display.index + 1

charges_outliers_display
```

## VISUALISASI NON OUTLIER

```python
sns.set_style('ticks')

fig, axes = plt.subplots(1,2, figsize=(5,8))

for ax, col in zip(axes, ['age', 'children']):
    sns.boxplot(
        y = df[col],
        ax = ax,
        color = 'navy'
    )
    ax.set_title(col.capitalize())

plt.tight_layout()
plt.show()
```

# VISUALISASI OUTLIER

```python
sns.set_style('white')
sns.set_context('talk')

features = ['bmi', 'charges']
colors = {
    'bmi': 'peru',       
    'charges': 'crimson'    
}

fig, axes = plt.subplots(len(features), 2, figsize=(12, 10))

for i, col in enumerate(features):
    # Boxplot
    sns.boxplot(
        y=df[col],
        ax=axes[i, 0],
        color=colors[col]
    )
    axes[i, 0].set_title(f'{col.capitalize()}  Boxplot')

    # Histogram + KDE
    sns.histplot(
        df[col],
        kde=True,
        bins=30,
        ax=axes[i, 1],
        color=colors[col]
    )
    axes[i, 1].set_title(f'{col.capitalize()}  Histogram')

plt.tight_layout()
plt.show()
```
